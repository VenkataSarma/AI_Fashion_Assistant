```python
user_profile = {
    "gender":"women",
    "occasion":"party",
    "wear_type":"western",
    "budget":3000
}
```

```python
def calculate_score(
    outfit,
    user_profile
):

    score = 0

    if outfit["gender"] == user_profile["gender"]:
        score += 30

    if outfit["occasion"] == user_profile["occasion"]:
        score += 30

    if outfit["wear_type"] == user_profile["wear_type"]:
        score += 20

    budget = user_profile["budget"]

    price = outfit["total_price_inr"]

    if price <= budget:
        score += 20

    return score
```

```python
scores = []

for _, outfit in outfits.iterrows():

    scores.append(
        calculate_score(
            outfit,
            user_profile
        )
    )

outfits["score"] = scores
```

```python
ranked = outfits.sort_values(
    by="score",
    ascending=False
)

ranked.head(10)
```
