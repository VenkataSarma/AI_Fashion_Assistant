```python
import pandas as pd
import numpy as np

products = pd.read_csv("D:/Dare X AI/Data/products.csv")
outfits = pd.read_csv("D:/Dare X AI/Data/outfits.csv")
```

```python
user_profile = {
    "gender":"women",
    "occasion":"party",
    "wear_type":"western"
}
```

```python
def calculate_score(
    outfit,
    user_profile
):

    score = 0

    # Gender Match
    if outfit["gender"].lower() == user_profile["gender"].lower():
        score += 40

    # Occasion Match
    if outfit["occasion"].lower() == user_profile["occasion"].lower():
        score += 40

    # Wear Type Match
    if outfit["wear_type"].lower() == user_profile["wear_type"].lower():
        score += 20

    return score
```

```python
calculate_score(
    outfits.iloc[0],
    user_profile
)
```

```python
scores = []

for _, outfit in outfits.iterrows():

    score = calculate_score(
        outfit,
        user_profile
    )

    scores.append(score)

outfits["score"] = scores
```

```python
outfits[
[
    "theme",
    "occasion",
    "wear_type",
    "score"
]
].head()
```

```python
ranked_outfits = outfits.sort_values(
    by="score",
    ascending=False
)

ranked_outfits.head(10)
```

```python
def recommend_top_k(
    user_profile,
    k=5
):

    scores = []

    for _, outfit in outfits.iterrows():

        score = calculate_score(
            outfit,
            user_profile
        )

        scores.append(score)

    temp = outfits.copy()

    temp["score"] = scores

    temp = temp.sort_values(
        by="score",
        ascending=False
    )

    return temp.head(k)
```

```python
recommend_top_k(
    {
        "gender":"women",
        "occasion":"party",
        "wear_type":"western"
    }
)
```

```python
top5 = recommend_top_k(
    {
        "gender":"women",
        "occasion":"party",
        "wear_type":"western"
    }
)

top5[
[
    "theme",
    "hero",
    "footwear",
    "score"
]
]
```
