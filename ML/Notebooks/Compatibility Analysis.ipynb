```python
compatibility_pairs = []

for _, row in outfits.iterrows():

    hero = row["hero"]

    if pd.notna(row["footwear"]):

        compatibility_pairs.append(
            (hero,row["footwear"])
        )

    if pd.notna(row["accessory_1"]):

        compatibility_pairs.append(
            (hero,row["accessory_1"])
        )

    if pd.notna(row["accessory_2"]):

        compatibility_pairs.append(
            (hero,row["accessory_2"])
        )

compatibility_pairs[:20]
```

```python
compatibility_df = pd.DataFrame(
    compatibility_pairs,
    columns=["hero","compatible_item"]
)

compatibility_df.head()
```

```python
compatibility_dict = {}

for hero,item in compatibility_pairs:

    if hero not in compatibility_dict:
        compatibility_dict[hero] = []

    compatibility_dict[hero].append(item)
```

```python
hero = list(compatibility_dict.keys())[0]

print(hero)

print(compatibility_dict[hero])
```
