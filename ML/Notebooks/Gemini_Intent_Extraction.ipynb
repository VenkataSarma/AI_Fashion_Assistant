````python
SYSTEM_PROMPT = """
You are an intent extraction system.

Extract:

gender
occasion
wear_type
budget

Return ONLY valid JSON.

Possible genders:
men
women

Possible occasions:
casual
office
party
wedding
vacation
sports

Possible wear types:
western
ethnic

Example:

Input:
I am a 22 year old male attending a wedding

Output:
{
 "gender":"men",
 "occasion":"wedding",
 "wear_type":"ethnic",
 "budget":null
}
"""
````

````python
def extract_intent(query):

    response = model.generate_content(
        SYSTEM_PROMPT + "\n\nUser Query:\n" + query
    )

    text = response.text.strip()

    if text.startswith("```json"):
        text = text.replace("```json","")
        text = text.replace("```","")

    return json.loads(text)
````

````python
query = "I am a 22 year old male attending a wedding"

intent = extract_intent(query)

intent
````
