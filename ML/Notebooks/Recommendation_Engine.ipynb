```python
import pandas as pd

products = pd.read_csv("D:/Dare X AI/Data/products.csv")
outfits = pd.read_csv("D:/Dare X AI/Data/outfits.csv")

print(products.shape)
print(outfits.shape)
```

```python
def recommend_outfits(
    gender,
    occasion
):

    results = outfits[
        (outfits["gender"].str.lower()==gender.lower())
        &
        (outfits["occasion"].str.lower()==occasion.lower())
    ]

    return results
```

```python
recommend_outfits(
    "women",
    "party"
)
```

```python
def recommend_best_outfit(
    gender,
    occasion
):

    filtered = outfits[
        (outfits["gender"].str.lower()==gender.lower())
        &
        (outfits["occasion"].str.lower()==occasion.lower())
    ]

    if len(filtered)==0:
        return None

    return filtered.iloc[0]
```

```python
result = recommend_best_outfit(
    "women",
    "party"
)

result
```

```python
def display_outfit(row):

    print("="*50)

    print("Theme:", row["theme"])

    print("\nHero:")
    print(row["hero"])

    print("\nFootwear:")
    print(row["footwear"])

    print("\nAccessory 1:")
    print(row["accessory_1"])

    print("\nAccessory 2:")
    print(row["accessory_2"])

    print("\nStylist Reasoning:")
    print(row["stylist_rationale"])
```

```python
def recommend_best_outfit(
    gender,
    occasion,
    wear_type=None
):

    filtered = outfits[
        (outfits["gender"].str.lower()==gender.lower())
        &
        (outfits["occasion"].str.lower()==occasion.lower())
    ]

    if wear_type:

        filtered = filtered[
            filtered["wear_type"].str.lower()
            ==
            wear_type.lower()
        ]

    if len(filtered)==0:
        return None

    return filtered.iloc[0]
```

```python
recommend_best_outfit(
    "women",
    "wedding",
    "ethnic"
)
```
